```python
import scipy.io as sio
import matplotlib.pyplot as plt
import numpy as np

# Load the .mat file
data_path = path+'/data/samson_dataset.mat'
data = sio.loadmat(data_path)
Y = data['Y']  # Hyperspectral image data
def plot_spectral_band(band_data, band_number):
    plt.figure()
    plt.plot(band_data)
    plt.title(f'Spectral Band {band_number}')
    plt.xlabel('Pixel Index')
    plt.ylabel('Reflectance')
    plt.show()

plot_spectral_band(Y[:, 0], 1)  #1
plot_spectral_band(Y[:, Y.shape[1] // 2], Y.shape[1] // 2) #mid
plot_spectral_band(Y[:, -1], Y.shape[1])  # Last ban#
```
